#1.Reverse of an array:

```python
def reversearray(arr, size):
    st = 0
    end = size - 1
    while st < end:
        temp = arr[st]
        arr[st] = arr[end]
        arr[end] = temp
        st += 1
        end -= 1

arr = [1, 2, 3, 4, 5, 6, 7]
size = len(arr)

for i in arr:
    print(i, end=" ")
print()

reversearray(arr, size)

for i in arr:
    print(i, end=" ")
```

#4. Find the Second Largest Element

```python
def secondlargest(arr):
    largest=arr[0];
    secondL=arr[0];
    for i in arr:
      if(i>largest):
        secondL=largest
        largest=i;
      elif(i>secondL and i<largest):
        secondL=i;
    return secondL;
arr=[1,2,3,4,5,6,7,8,9]
print(secondlargest(arr))
```

#5. Count Frequency of Elements


```python
def count_frequency(arr):
    freq = {}
    for i in arr:
        freq[i] = freq.get(i, 0) + 1
    return freq

arr = [1,2,2,3,3,3,4]
print(count_frequency(arr))
```

#6. Check if Array is Sorted

```python
def is_sorted(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            return False
    return True

arr = [1,2,3,4,5]
print(is_sorted(arr))
```

#7. Rotate Array Right by k


```python
def rotate_right(arr, k):
    k = k % len(arr)
    return arr[-k:] + arr[:-k]

arr = [1,2,3,4,5]
print(rotate_right(arr,2))
```

#8. Find Pair with Given Sum

```python
def find_pair(arr, target):
    s = set()
    for i in arr:
        if target - i in s:
            return (i, target - i)
        s.add(i)
    return None

arr = [1,2,3,4,5]
print(find_pair(arr,7))
```

#9. Remove Duplicates

```python
def remove_duplicates(arr):
    seen = set()
    result = []
    for i in arr:
        if i not in seen:
            result.append(i)
            seen.add(i)
    return result

arr = [1,2,2,3,4,4,5]
print(remove_duplicates(arr))
```

#12. Find Missing Number

```python
def find_missing(arr, n):
    return n*(n+1)//2 - sum(arr)

arr = [1,2,4,5]
print(find_missing(arr,5))
```

#13. Find Duplicates

```python
def find_duplicates(arr):
    seen = set()
    dup = set()
    for i in arr:
        if i in seen:
            dup.add(i)
        else:
            seen.add(i)
    return list(dup)

arr = [1,2,3,2,4,5,1]
print(find_duplicates(arr))
```

#17. Leader Elements

```python
def leaders(arr):
    result = []
    max_right = float('-inf')
    for i in reversed(arr):
        if i > max_right:
            result.append(i)
            max_right = i
    return result[::-1]

print(leaders([16,17,4,3,5,2]))
```

#18. Move Zeroes to End

```python
def move_zeroes(arr):
    non_zero = [i for i in arr if i != 0]
    zeroes = [0]*(len(arr)-len(non_zero))
    return non_zero + zeroes

print(move_zeroes([1,0,2,0,3,4]))
```

#19. Subarray with Given Sum

```python
def subarray_sum(arr, target):
    curr = 0
    start = 0
    for end in range(len(arr)):
        curr += arr[end]
        while curr > target and start <= end:
            curr -= arr[start]
            start += 1
        if curr == target:
            return arr[start:end+1]
    return None

print(subarray_sum([1,2,3,4,5],9))
```

#22. All Subarrays

```python
def all_subarrays(arr):
    result = []
    for i in range(len(arr)):
        for j in range(i,len(arr)):
            result.append(arr[i:j+1])
    return result

print(all_subarrays([1,2,3]))
```

#23. Maximum Sum Subarray

```python
def max_subarray(arr):
    max_sum = curr = arr[0]
    for i in arr[1:]:
        curr = max(i, curr+i)
        max_sum = max(max_sum, curr)
    return max_sum

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

#24. Rearrange Alternately

```python
def rearrange_alternate(arr):
    arr.sort()
    result = []
    i = 0
    j = len(arr)-1
    while i <= j:
        if i != j:
            result.append(arr[j])
            result.append(arr[i])
        else:
            result.append(arr[i])
        i += 1
        j -= 1
    return result

print(rearrange_alternate([1,2,3,4,5,6]))
```

#25. Majority Element

```python
def majority_element(arr):
    count = 0
    candidate = None
    for i in arr:
        if count == 0:
            candidate = i
        count += 1 if i == candidate else -1
    return candidate

print(majority_element([2,2,1,2,3,2,2]))
```

#26. Find Peak Element

```python
def find_peak(arr):
    for i in range(len(arr)):
        if (i==0 or arr[i]>arr[i-1]) and (i==len(arr)-1 or arr[i]>arr[i+1]):
            return arr[i]

print(find_peak([1,3,20,4,1,0]))
```

#27. First Missing Positive

```python
def first_missing_positive(arr):
    s = set(arr)
    i = 1
    while True:
        if i not in s:
            return i
        i += 1

print(first_missing_positive([3,4,-1,1]))
```

#28. Sort 0s, 1s, 2s

```python
def sort_012(arr):
    count0 = arr.count(0)
    count1 = arr.count(1)
    count2 = arr.count(2)
    return [0]*count0 + [1]*count1 + [2]*count2

print(sort_012([0,1,2,0,1,2]))
```

#29. Longest Consecutive Sequence

```python
def longest_consecutive(arr):
    s = set(arr)
    longest = 0
    for i in s:
        if i-1 not in s:
            length = 1
            while i+length in s:
                length += 1
            longest = max(longest,length)
    return longest

print(longest_consecutive([100,4,200,1,3,2]))
```

#30. Product of Array Except Self

```python
def product_except_self(arr):
    n = len(arr)
    left = [1]*n
    right = [1]*n
    for i in range(1,n):
        left[i] = left[i-1]*arr[i-1]
    for i in range(n-2,-1,-1):
        right[i] = right[i+1]*arr[i+1]
    return [left[i]*right[i] for i in range(n)]

print(product_except_self([1,2,3,4]))
```
